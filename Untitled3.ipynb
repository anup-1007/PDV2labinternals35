```python
import pandas as pd 

df = pd.read_csv("milk_production.csv")
df.head()
```

```python
import pandas as pd 


df["maxmilk2013_14"] = df["Cow Milk-2013-14"] + df["Boffalo Milk-2013-14"] + df["Goat Milk-2013-14"]

df[df["maxmilk2013_14"]==df["maxmilk2013_14"].max()]["State/ UT Name"]
```

```python
df["y2011"]=df["Cow Milk-2010-11"] + df["Boffalo Milk-2010-11"] + df["Goat Milk-2010-11"]
df["y2012"]=df["Cow Milk-2011-12"] + df["Boffalo Milk-2011-12"] + df["Goat Milk-2011-12"]
df["y2014"]=df["Cow Milk-2013-14"] + df["Boffalo Milk-2013-14"] + df["Goat Milk-2013-14"]
df["y2015"]=df["Cow Milk-2014-15"] + df["Boffalo Milk-2014-15"] + df["Goat Milk-2014-15"]
df["y2016"]=df["Cow Milk-2015-16"] + df["Boffalo Milk-2015-16"] + df["Goat Milk-2015-16"]

print("2011 top 5states")
print(df.sort_values("y2011",ascending=False)["State/ UT Name"].head())
print("2012 top 5states")
print(df.sort_values("y2012",ascending=False)["State/ UT Name"].head())
print("2014 top 5states")
print(df.sort_values("y2014",ascending=False)["State/ UT Name"].head())
print("2015 top 5states")
print(df.sort_values("y2015",ascending=False)["State/ UT Name"].head())
print("2016 top 5states")
print(df.sort_values("y2016",ascending=False)["State/ UT Name"].head())
```

```python
df["year 2011"]= df["Cow Milk-2010-11"]+df["Boffalo Milk-2010-11"]+df["Goat Milk-2010-11"]
print("2011 top 5 states")
print(df.sort_values("year 2011",ascending = False).head())
```

```python
df[df["y2016"].ge(df["y2015"]) & df["y2015"].ge(df["y2014"])].head()
```

```python
def find_avg(milk_name):
    ans=df.loc[:,"y2011":"y2016"]
    ans=ans.sum(axis=1)/5
    ans=pd.DataFrame(ans)
    ans["State/ UT Name"]=df["State/ UT Name"]
    
    ans=ans[ans["State/ UT Name"]==milk_name]
    
    
    return ans[0].values
print(find_avg("Assam"))









```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x= ["y2011","y2012","y2013","y2014","y2015"]
y = [df.get_value(0,"y2011"),df.get_value(0,"y2012"),df.get_value(0,"y2014"),df.get_value(0,"y2015"),df.get_value(0,"y2016")]
plt.plot(x,y,label="line graph for Andra Pradesh",marker="o")

plt.xlabel("last 5years")
plt.ylabel("milk production")
plt.legend()

plt.show()
```

```python

cow = [df.get_value(0,"Cow Milk-2011-12"),df.get_value(0,"Cow Milk-2013-14"),df.get_value(0,"Cow Milk-2014-15"),df.get_value(0,"Cow Milk-2015-16")]
boffalo = [df.get_value(0,"Boffalo Milk-2011-12"),df.get_value(0,"Boffalo Milk-2013-14"),df.get_value(0,"Boffalo Milk-2014-15"),df.get_value(0,"Boffalo Milk-2015-16")]
goat =[df.get_value(0,"Goat Milk-2011-12"),df.get_value(0,"Goat Milk-2013-14"),df.get_value(0,"Goat Milk-2014-15"),df.get_value(0,"Goat Milk-2015-16")]

fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(wspace = 0.5, hspace = 0.3)

axes[0,0].pie(cow,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[0,0].set_title("cow production")

axes[0,1].pie(boffalo,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[0,1].set_title("Boffalo production")

axes[1,0].pie(goat,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[1,0].set_title("Goat production")
plt.show()
```
